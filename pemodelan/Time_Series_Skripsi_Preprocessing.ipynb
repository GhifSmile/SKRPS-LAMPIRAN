```python
# Melakukan konversi untuk tanggal pada data menjadi format %d %B %Y

def convert_date(date_string):
  month_mapping = {
    'Jan': 'January',
    'Feb': 'February',
    'Mar': 'March',
    'Apr': 'April',
    'Mei': 'May',
    'Jun': 'June',
    'Jul': 'July',
    'Agu': 'August',
    'Sep': 'September',
    'Okt': 'October',
    'Nov': 'November',
    'Des': 'December'
}
  for ind_month, eng_month in month_mapping.items():
      date_string = date_string.replace(ind_month, eng_month)

  date_string = date_string.strip().split(', ')[1]

  return pd.to_datetime(' '.join(date_string.split()[0:-1]), format='%d %B %Y')
```

```python
df['date'] = df['date'].apply(lambda x: convert_date(x)) # Apply function convert_date per data date
```

## Text Preprocessing

### one-on-one processing

This section is just to explain how to process each part individually.

```python
import re

def coba_cleansing(text):
  text = text.replace('\xa0', ' ')
  text = text.replace('Pasardana.id –', '')
  text = text.replace('Pasardana.id -', '')
  text = re.sub(r' +', ' ', text)

  return text
```

```python
df['cleansing'] = df.article.apply(lambda x: coba_cleansing(x))
```

```python
df['case_folding'] = df.cleansing.apply(lambda x: x.lower())
```

```python
def pemilahan (df):

  result = []

  for i in range(len(df)):

    list_text = sentiment_prepro(df['article'].iloc[i])

    if 'lq45' not in df['title'].iloc[i].lower():
      list_text = [text.lower() for text in list_text if 'lq45' in text.lower()]

    else:
      list_text = [text.lower() for text in list_text]

    result.append(list_text)

  return result
```

```python
pemilahan_kata = pemilahan(df)
```

```python
import pandas as pd

df_pemilahan = pd.DataFrame({'pemilahan': pemilahan_kata})
```

```python
df['pemilahan'] = df_pemilahan['pemilahan']
```

```python
from textblob import TextBlob
from textblob.exceptions import NotTranslated

def trlst(df):

  translation_result = []

  for i in range(len(df)):

    result = []

    list_text = df.pemilahan.iloc[i]

    for sentence in list_text:

      try:
        sentence = sentence.lower()
        blob = TextBlob(sentence.strip())
        translation = blob.translate(from_lang='id',to='en')
        result.append(str(translation.lower()))

      except NotTranslated:
        continue

    translation_result.append(result)

  return translation_result
```

```python
trlst_df = trlst(df)
```

```python
df_translasi = pd.DataFrame({'translasi': trlst_df})
```

```python
df['translasi'] = df_translasi['translasi']
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

def sentimen_score(df):

  list_sentiment_score = []
  mean_sentimen_score = []

  sid_obj = SentimentIntensityAnalyzer()

  for i in range(len(df)):

    sentimen_score = []

    list_text = df.translasi.iloc[i]

    for sentence in list_text:

        sentiment_dict = sid_obj.polarity_scores(sentence)
        sentimen_score.append(sentiment_dict['compound'])

    list_sentiment_score.append(sentimen_score)
    mean_sentimen_score.append(np.mean(sentimen_score))

  return pd.DataFrame({'list_sentiment_score': list_sentiment_score, 'mean_sentiment_score':mean_sentimen_score})
```

```python
sentimen = sentimen_score(df)
```

### accumulation processing

```python
def sentiment_prepro(text):

  text = text.replace('\xa0', ' ')
  text = text.replace('Pasardana.id –', '')
  text = text.replace('Pasardana.id -', '')

  text = text.strip()

  return text.split('\n')
```

```python
from textblob import TextBlob
from textblob.exceptions import NotTranslated
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

def sentiment_scores(df):
  val = list(np.zeros(len(df)))

  sid_obj = SentimentIntensityAnalyzer()

  for i in range(len(df)):

    text_sentiment = []

    list_text = sentiment_prepro(df['article'].iloc[i])

    # Mengambil kalimat yang mengandung kata 'lq45', jika pada title tidak mengandung kata 'lq45'
    if 'lq45' not in df['title'].iloc[i].lower():
      list_text = [text for text in list_text if 'lq45' in text.lower()]

    for sentence in list_text:

      try:
        sentence = sentence.lower()
        blob = TextBlob(sentence.strip())
        translation = blob.translate(from_lang='id',to='en')
        sentiment_dict = sid_obj.polarity_scores(str(translation.lower()))
        text_sentiment.append(sentiment_dict['compound'])

      except NotTranslated:
        continue

    scores = np.mean(text_sentiment) # Mengambil rata-rata sentiment score untuk setiap article
    val[i] = scores

  return val
```

```python
df['sentiment_article'] = sentiment_scores(df)
```

```python
# Mendapatkan rata-rata sentiment score per tanggal publikasi
def sentiment_merge(df):

  val = list(np.zeros(len(df.index.unique())))

  for i, j in enumerate(df.index.unique()):

    idx = j
    open = np.mean(df['Open'].loc[j])
    high = np.mean(df['High'].loc[j])
    low = np.mean(df['Low'].loc[j])
    close = np.mean(df['Close'].loc[j])
    sentiment_article = np.mean(df['sentiment_article'].loc[j])

    val[i] = [idx, open, high, low, close, sentiment_article]

  return pd.DataFrame(val, columns=['date', 'open', 'high', 'low', 'close', 'sentiment_score'])
```
